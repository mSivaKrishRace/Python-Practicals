#### Merging all the Different Sections files into Single Dataframe

```python
Apex21BDA = pd.concat([Apex21BDA1,Apex21BDA2,Apex21BDA3], ignore_index=True)
```

```python
Apex21BDA = Apex21BDA.rename(columns={"Student Name" : "Name"})
```

```python
Apex21BDA = Apex21BDA.rename(columns={'Student UID':'Stud_UID'})
```

```python
Apex21BDA.head()
```

#### Below Conversion of Strings to any Common case is mandetory for proper comparision

```python
Apex21BDA["Name"]= Apex21BDA["Name"].str.upper()
Non_RegisteredStuds["Name"] = Non_RegisteredStuds["Name"].str.upper()
```

#### performing INNER_JOIN on the both Dataframes

```python
New_Studs_Contact_list = Apex21BDA.merge(Non_RegisteredStuds, left_on="Name", right_on="Name", how="right")
```
